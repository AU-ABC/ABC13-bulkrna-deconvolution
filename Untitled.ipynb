```r
library(Seurat)
```

```r
scrnaMatrix <- ReadMtx("Data/GSE109033_matrix.mtx.gz",
       "Data/GSE109033_barcodes.tsv.gz",
       "Data/GSE109033_genes.tsv.gz",)
```

```r
scrnaData <- CreateSeuratObject(counts = scrnaMatrix, project = "mouse_testis")
```

```r
samples <-sapply( colnames(scrnaData), function(x) unlist(strsplit(x,"-"))[2] )  
```

```r
scrnaData@meta.data['Sample'] = samples
```

```r
int.list <- list(scrnaData1, 
                  scrnaData2, 
                  scrnaData3)
```

```r
int.list <- lapply(X = int.list, FUN = function(x) {
  message("Normalizing\n")
  x <- SCTransform(x, ncells=3000, vst.flavor = 'v1' , vars.to.regress = c("percent.mt"), variable.features.n = 2000, return.only.var.genes = FALSE, verbose=FALSE)
})
```

```r
int.features <- SelectIntegrationFeatures(object.list = int.list, nfeatures = 2000)

int.list <- PrepSCTIntegration(object.list = int.list, anchor.features = int.features)
```

```r
int.anchors <- FindIntegrationAnchors(object.list = int.list, normalization.method = "SCT", 
                                       anchor.features = int.features, reference = c(3))

scrnaInt <- IntegrateData(anchorset = int.anchors, normalization.method = "SCT")
```

```r
cat("The default assay of the data is now called: ")
cat(DefaultAssay(scrnaInt))
```

```r
scrnaInt <- RunPCA(object = scrnaInt, verbose = FALSE)
```

```r
ElbowPlot(scrnaInt, ndims = 30)
```

```r
scrnaInt <- FindNeighbors(object = scrnaInt, dims = 1:15, k.param = 30)
```

```r
scrnaInt <- RunUMAP(object = scrnaInt, dims = 1:20)
```

```r
scrnaInt <- SetIdent(scrnaInt, value = "Sample")
```
